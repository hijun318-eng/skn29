```python
import requests
from bs4 import BeautifulSoup
url = 'https://books.toscrape.com/'
response = requests.get(url)
response.raise_for_status() # 에러 체크

soup = BeautifulSoup(response.text, 'lxml')
titles, links = [], []
for li in soup.select('ol.row li'):
    titles.append(li.select('article.product_pod')[0].select_one('h3 a').get('title'))
    links.append(url + li.select('article.product_pod')[0].select_one('h3 a').get('href'))
```

```python
import pandas as pd
data = {
    'title': titles,
    'link': links
}
pd.DataFrame(data)
```

```python
import requests
from bs4 import BeautifulSoup

titles, links = [], []
for pagenum in range(1, 51):
    url = 'https://books.toscrape.com/catalogue/page-{pagenum}.html'
    response = requests.get(url)
    response.raise_for_status() # 에러 체크

    soup = BeautifulSoup(response.text, 'lxml')

    for li in soup.select('ol.row li'):
        titles.append(li.select('article.product_pod')[0].select_one('h3 a').get('title'))
        links.append('https://books.toscrape.com/' + li.select('article.product_pod')[0].select_one('h3 a').get('href'))
```

```python
import pandas as pd
data = {
    'title': titles,
    'link': links
}
pd.DataFrame(data)
```
